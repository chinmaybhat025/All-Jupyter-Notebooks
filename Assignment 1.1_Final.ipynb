## Bar Graph:-
## A bar chart or bar graph is a chart or graph that presents categorical data with rectangular bars with heights or lengths proportional to the values that they represent. The bars can be plotted vertically or horizontally. A vertical bar chart is sometimes called a line graph.However there are different types of Bar Graph. 

## 1) Simple Bar chart
## Simple bar graphs compare data with one independent variable and can relate to a set point or range of data. 

## Example:- Bar Graph Consists of different branches of Engg(X-axis)  who are pursuing the engineering field  based on the packages allotted for each field by the CEO's of each organization.

```python
import pandas as pd # Importing the Panda's directory
import matplotlib
import matplotlib.pyplot as plt 
%matplotlib inline

#create a list of different fields of Engineers in the country
Engg=['CHEMICAL','COMPUTER','MECHANICAL','CIVIL','AUTOMOBILE','PRODUCTION'] 

# create a list of different packages alloted by the CEO's of each organization
Salary=[120000,600000,200000,110000,100000,120000]

# Pandas Series is a one-dimensional labeled array capable of holding data of any type (integer, string, float, python objects, etc.)
# creating a dictionary
d={'Engineering': pd.Series(Engg),'Salary': pd.Series(Salary)} #reference:- https://www.geeksforgeeks.org/python-pandas-series/

# create a pandas dataframe from dictionary
dframe=pd.DataFrame(d) # reference:-# https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.html
plt.bar(['CHEM','COMP','MECH','CIVIL','AUTO','PROD'],[120000,500000,200000,110000,100000,120000], label='Engg fields vs Packages',color='b')
plt.xlabel('Engg Fields')
plt.ylabel('Packages')
plt.title("Different Engg fields offering Estimated packages for the Upcoming Freshers who are graduated in Year 2019 ")
plt.legend()
#plt.xticks('Engg Fields', fontsize=30)
#plt.rcParams['figure.figsize']=(3,4) 
plt.bar(5,5)
```

## Observations from above simple bar chart:-
## i) Based on the above data, we can see that student's who are pursuing the computer engineering gets the highest package than any other branch.
## ii) However, Student's who are pursuing Automobile Engineering gets the lowest package.
## iii) Student's who are pursuing/about to pursue Mechanical Engineering
## get's the second highest package after Computer Engineering.

##  2) Multiple Bar chart
## A multiple bar graph is a graph that has multiple bars for each category. Multiple bar graphs are designed to compare the data collected during multiple times or events.

## Example:- Different shops of cold-drinks producing quantities of pepsi, coca-cola and Thumps up weekly analysis

```python
import numpy as np
import pandas as pd
import random as rd # when we want to pick the random number from range
import matplotlib.pyplot as plt

rd.seed(12) # This is the seed for the next random number for which random directory from the range
cold_drinks=[
    rd.sample((range(1,20)),5), # reference:-https://docs.python.org/2/library/random.html
    rd.sample((range(1,20)),5),
    rd.sample((range(1,20)),5)
]
cold_drinks
x=np.arange(5) # reference:-https://docs.scipy.org/doc/numpy/reference/generated/numpy.arange.html
days="Mon","Tue","Wed","Thu","Fri"
plt.bar(x+0.00, cold_drinks[0], color="b", width=0.25, Label="PEPSI")# #reference:-https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.bar.html.
plt.bar(x+0.25, cold_drinks[1], color="r", width=0.25, Label="COCA-COLA")
plt.bar(x+0.50, cold_drinks[2], color="g", width=0.25, Label="THUMPS-UP")
plt.title("COLD DRINKS",fontsize=10) # reference:-https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.title.html
plt.xticks(x,days) # reference:-https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.xticks.html
plt.legend(fontsize=20, loc=(1.05,0.4)) # reference:-https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.legend.html
plt.show() # reference:-https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.show.html
```

## Observations from the Multiple Bar chart
## i) Based on the data, we have found that, on thursday, Maximum quantity of Pepsi has been produced.
## ii) Based on the data, we have found that, on Friday, Maximum quantity of Thumps-up has been produced.
## iii) Based on the data, we have found that, on thursday, Maximum quantity of Coca-Cola ALSO has been produced.
## iv) Based on the data, we have found that, on Friday, Minimum quantity of Pepsi has been produced.
## v) Based on the data, we have found that, on Wednesday, Maximum quantity of Thumps-up has been produced.
## vi) Based on the data, we have found that, on Tuesday, Maximum quantity of Coca-cola has been produced.

## 3) Stack Bar Chart

## A stacked bar graph (or stacked bar chart) is a chart that uses bars to show comparisons between categories of data, but with ability to break down and compare parts of a whole

## Example:- Weather forecast for mumbai for the whole week analysis

```python
import numpy as np
import pandas as pd
import random as rd # when we want to pick the random number from range
import matplotlib.pyplot as plt

rd.seed(12) # This is the seed for the next random number for which random directory from the range
Weather=[
    rd.sample((range(1,20)),5),# reference:-https://docs.python.org/2/library/random.html
    rd.sample((range(1,20)),5),
    rd.sample((range(1,20)),5)
]
Weather
x=np.arange(5) # reference:-https://docs.scipy.org/doc/numpy/reference/generated/numpy.arange.html
days="Mon","Tue","Wed","Thu","Fri"
plt.bar(x, Weather[0], bottom=np.sum(Weather[:0],axis=0), label="RAIN")#reference:-https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.bar.html.
plt.bar(x, Weather[1], bottom=np.sum(Weather[:1], axis=0), label="SUMMER")
plt.bar(x, Weather[2], bottom=np.sum(Weather[:2],axis=0), label="WINTER")
plt.title("Weather",fontsize=10) # https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.title.html
plt.xticks(x,days) #reference:-https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.xticks.html
plt.legend(fontsize=20, loc=(1.05,0.4)) # reference:-https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.legend.html
plt.show() # reference:-https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.show.html
```

## Observations from the stacked bar chart:-
## i) Based on the above data, Wednesday, maximum part of the whole day has rain, however, on Friday, there was very neglibile amount of rain as compared to the Wednesday.
## ii) On Thursday, it has been found that, approximately equal part of the rain and summery day has been found.
## iii) on Friday, it has been found that, Maximum part of the day is Winter.

## 4)Percentage Bar chart
## Percentage bar graphs compare the percentage that each item contributes to an entire category. Rather than showing the data as clusters of individual bars, percentage bar graphs show a single bar with each measured item represented by a different color.

##  Example:- Below showing the Poverty rate of three countries(India,Bangladesh,Pakistan) from the 2014 to 2019. 

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
import pandas as pd
# Data
row = [0,1,2,3,4]
raw_data = {'India': [20, 1.5, 7, 10, 5], 'Pakistan': [5, 15, 5, 10, 15],'Bangladesh': [2, 15, 18, 5, 10]}
df = pd.DataFrame(raw_data)

# From raw value to percentage
totals = [i+j+k for i,j,k in zip(df['India'], df['Pakistan'], df['Bangladesh'])]
India = [i / j * 100 for i,j in zip(df['India'], totals)]
Pakistan = [i / j * 100 for i,j in zip(df['Pakistan'], totals)]
Bangladesh = [i / j * 100 for i,j in zip(df['Bangladesh'], totals)]

# plot
barWidth = 0.90
year = (2014,2015,2016,2017,2018)
# Create Bars for India
plt.bar(row, India, color='#b5ffb9', edgecolor='white', width=barWidth,label='India poverty rate')
# Create Bars for Pakistan
plt.bar(row, Pakistan, bottom=India, color='#f9bc86', edgecolor='white', width=barWidth,label='Pakistan poverty rate')
# Create Bars for Bangladesh
plt.bar(row, Bangladesh, bottom=[i+j for i,j in zip(India, Pakistan)], color='#a3acff', edgecolor='white', width=barWidth,label='Bangladesh Poverty rate')

# Custom x axis
plt.xticks(row, year)
plt.xlabel(" Year")
plt.ylabel("Percentages")
plt.legend(loc='upper center')
plt.title("Poverty rate for three countries from year 2014 to 2019")

# Show graphic
plt.show()
```

## Observations from above percentage bar chart:-
## i) Based on the above data, India's Poverty rate has been approximately to 70% which is most than the Pakistan and Bangladesh's poverty rate combined.
## ii) However, over the 1 year passed, India's Poverty rate has been least as compared to the year 2014.
## iii) After Year 2015 till the year 2018, there is significant increase and decrease in the India's Poverty rate, which has been the worrying signs for the country.

## 6) Histogram
## The major difference is that a histogram is only used to plot the frequency of score occurrences in a continuous data set that has been divided into classes, called bins.

## Example:-using Histogram, we will able to find out how many students passed with distinction, first class and second class

```python
import matplotlib.pyplot as plt
%matplotlib inline
#Marks=[60,75,55,61,63,80,73,89,59] 
Boys_Marks=[60,75,80,90,55,50,89,77]
Girls_Marks=[77,70,78,85,84,55,50,99]
plt.xlabel('Marks range')
plt.ylabel('total number of students getting those marks')
plt.title('Marks analysis for 10th Students after SSC exam 2018-19')
plt.hist([Boys_Marks,Girls_Marks], bins=[50,60,75,90],rwidth=0.95,color=['blue','green'], label=['Boys','Girls']) #https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.hist.html
plt.legend()
```

## 8)Multiple Line chart
## A multiple line graph shows the relationship between independent and dependent values of multiple sets of data. Usually multiple line graphs are used to show trends over time.

## Example Showing the deaths happened from 2014 to 2019 due to Demonetisation

```python
import matplotlib.pyplot as plt
%matplotlib inline


year = [2014, 2015, 2016, 2017, 2018, 2019]
Haryana_deathrate = [45, 58, 78, 107, 138, 170]
Maharashtra_deathrate = [449, 553, 696, 870, 100, 130]
plt.plot(year, Haryana_deathrate, color='g',label="Haryana")
plt.plot(year, Maharashtra_deathrate, color='orange',label="Maharashtra")
plt.xlabel('Year')
plt.ylabel('Dead')
plt.title('Death rate in Two States')
plt.legend()
plt.show()
```
